# Student Grade Manager
This program helps a teacher manage student grades using file operations.

Features:

Add a student and their grade.

View all students and grades.

Calculate the average grade.

Exit the program.

```python
def add_student(filename):
    """Add a new student and their grade."""
    try:
        name = input("Enter student name: ")
        grade = float(input("Enter student grade: "))
        with open(filename, "a") as file:
            file.write(f"{name},{grade}\n")
        print("Student grade added successfully.")
    except ValueError:
        print("Error: Invalid grade. Please enter a number.")
    except Exception as e:
        print(f"An error occurred: {e}")


def view_students(filename):
    """View all students and their grades."""
    try:
        with open(filename, "r") as file:
            print("\nStudents and Grades:")
            for line in file:
                name, grade = line.strip().split(",")
                print(f"Name: {name}, Grade: {grade}")
    except FileNotFoundError:
        print("Error: Grade file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


def calculate_average(filename):
    """Calculate the average grade of all students."""
    try:
        with open(filename, "r") as file:
            total = 0
            count = 0
            for line in file:
                name, grade = line.strip().split(",")
                total += float(grade)
                count += 1
            if count > 0:
                print(f"Average Grade: {total / count:.2f}")
            else:
                print("No students found to calculate average.")
    except FileNotFoundError:
        print("Error: Grade file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


def grade_manager_menu():
    """Menu for student grade manager operations."""
    filename = "grades.txt"
    while True:
        print("\nGrade Manager Menu:")
        print("1. Add Student Grade")
        print("2. View Students and Grades")
        print("3. Calculate Average Grade")
        print("4. Exit")
        try:
            choice = int(input("Enter your choice (1-4): "))
            if choice == 1:
                add_student(filename)
            elif choice == 2:
                view_students(filename)
            elif choice == 3:
                calculate_average(filename)
            elif choice == 4:
                print("Exiting Grade Manager.")
                break
            else:
                print("Invalid choice. Please choose between 1 and 4.")
        except ValueError:
            print("Error: Please enter a valid integer choice.")


# Run the Grade Manager Program
grade_manager_menu()
```
